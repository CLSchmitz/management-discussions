```python
import numpy as np
import pandas as pd
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Population size
N = 1000

# Initial conditions: 1 Infected, no recovered, no deaths
I0, R0, D0 = 1, 0, 0 

# All others, S=N-I-R-D susceptible, initially S0=N-I0-R0
S0 = N - I0 - R0 - D0 #normalization

# Contact rate, beta
beta = 1

# A grid of time points (in days)
t = np.linspace(0, 150, 151)
```

# (a) SIRD Model Equations

```python
# SIRD model (coupled nonlinear system of ODEs)
def deriv(y, t, N, beta, gamma, mu):
    S, I, R, D = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - (gamma + mu)*I
    dRdt = gamma * I
    dDdt = mu * I
    return dSdt, dIdt, dRdt, dDdt

# Initial conditions vector
y0 = S0, I0, R0, D0
```

```python
### Part (b)

# Empty df for results
res = pd.DataFrame(columns = ['Reproduction Rate', 'Death Rate', 'Deaths', 'Total Infections'])

for r_nought in np.linspace(1.1, 4, 10): #loop through r_noughts
    for mu in np.linspace(0.001, 0.04, 10): #loop through mu
        
        # Calculate Gamma
        gamma = (1/r_nought) - mu
        
        #Integrate
        S, I, R, D = odeint(deriv, y0, t, args=(N, beta, gamma, mu)).T
        
        #Total Infections
        total_inf = 1000 - S[-1]

        #Append number dead and total infections after 1000 days to df
        res = res.append({'Reproduction Rate': r_nought.round(2),
                          'Death Rate': mu.round(4),
                          'Deaths': D[-1].round(0),
                          'Total Infections': total_inf},
                          ignore_index = True)

```

```python
#Infections by Reproduction Rate (Deaths Constant)
res_infections = res.drop_duplicates(subset=['Reproduction Rate'], keep='first')
sns.lineplot(data = res_infections, x = 'Reproduction Rate', y = 'Total Infections')
```

```python
#Deaths by mu (R0 Constant)
res_deaths = res[res['Reproduction Rate'] == 2.07]
sns.lineplot(data = res_deaths, x = 'Death Rate', y = 'Deaths')
```

```python
#Deaths by Reproduction and Death Rates
res_pivot = res.pivot(index = 'Reproduction Rate', 
                      columns = 'Death Rate', 
                      values = 'Deaths').iloc[::-1,:]
sns.heatmap(res_pivot,cbar_kws={'label': 'Deaths per 1000'})
```

```python
### Part (c): Lockdowns

#Updated deriv to allow vector of betas 
def deriv_beta_var(y, t, N, beta, gamma, mu, lockdown_t, lockdown_beta):
    S, I, R, D = y
    if t > lockdown_t: beta = lockdown_beta
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - (gamma + mu)*I
    dRdt = gamma * I
    dDdt = mu * I
    return dSdt, dIdt, dRdt, dDdt

#Function to plot developments in lockdown
def lockdown_plot(t, beta, gamma, mu, 
                  lockdown_t = None, 
                  lockdown_beta = None, 
                  plot_vars = ['S', 'I', 'R', 'D'], 
                  fig_name = None):
    
    #Baseline Model
    S, I, R, D = odeint(deriv, y0, t, args=(N, beta, gamma, mu)).T

    df = pd.DataFrame()
    df['Day'] = t
    df['S'] = S/10
    df['I'] = I/10
    df['R'] = R/10
    df['D'] = D/10

    #Lockdown Model
    if lockdown_t is not None:
        S_l, I_l, R_l, D_l = odeint(deriv_beta_var, y0, t, args=(N, beta, gamma, mu, lockdown_t, lockdown_beta)).T

        df['S_L'] = S_l/10
        df['I_L'] = I_l/10
        df['R_L'] = R_l/10
        df['D_L'] = D_l/10

    #Plot
    df = df.melt(id_vars = 'Day', value_vars=plot_vars)
    df = df.replace({'variable':{'S': 'Susceptible',
                                 'I': 'Infected',
                                 'R': 'Recovered',
                                 'D': 'Dead',
                                 'S_L': 'Susceptible (Lockdown)',
                                 'I_L': 'Infected (Lockdown)',
                                 'R_L': 'Recovered (Lockdown)',
                                 'D_L': 'Dead (Lockdown)'}})
    ax = sns.lineplot(data = df, x = 'Day', y = 'value', hue = 'variable')
    ax.set(xlabel='Days', ylabel='Percentage of Population')
    if lockdown_t is not None:
        ax.axvline(lockdown_t, linestyle = '--', lw=0.7)
        ax.text(lockdown_t +1 ,df['value'].max()*0.6,'Lockdown',rotation=0)
        ax.set_title('Lockdown from beta {} to beta {} after {} days'.format(beta, lockdown_beta,lockdown_t))
    else: ax.set_title('Model without Lockdown')
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles=handles[1:], labels=labels[1:], title="Group",loc = 'upper right')
    
    #Save (if desired)
    if fig_name is not None:
        ax.figure.savefig(fig_name +'.png')
    return df
```

```python
#Baseline model
df_0 = lockdown_plot(t, 0.25, 0.15, 0.02, fig_name = 'no_lockdown')
```

```python
df_1 = lockdown_plot(t, 0.25, 0.15, 0.02, 25, 0.15, ['I','D','I_L','D_L'], 'strong_early') #Strong Early
```

```python
df_2 = lockdown_plot(t, 0.25, 0.15, 0.02, 60, 0.15, ['I','D','I_L','D_L'], 'strong_late') #Strong Late
```

```python
res = pd.DataFrame()
for lockdown_beta in np.linspace(0.1, 0.24, 15):
    for lockdown_t in np.linspace(10,80,15):
        
        S, I, R, D = odeint(deriv_beta_var, y0, t, args=(N, 0.25, 0.15, 0.02, lockdown_t, lockdown_beta)).T
        deaths = D[-1]/10
        res = res.append({'Lockdown Beta': lockdown_beta.round(2),
                          'Lockdown Time': lockdown_t.round(0),
                          'Deaths': deaths}, ignore_index=True)

res_pivot = res.pivot(index = 'Lockdown Beta', 
                      columns = 'Lockdown Time', 
                      values = 'Deaths').iloc[::-1,:]
sns.heatmap(res_pivot,cbar_kws={'label': 'Deaths in Percent of Population'})

```
